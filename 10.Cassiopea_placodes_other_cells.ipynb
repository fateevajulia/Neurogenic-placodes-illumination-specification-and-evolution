```python
data1=data.copy()
```

```python
data1
```

```python
# Set the 'TF' column as the index
data1.set_index('TF', inplace=True)

```

```python
data1
```

```python
data1=data1[(data1.sum(axis=1) > 0) & (data1.sum(axis=1) < 5) ]
data1
```

```python
data1=data1.T
```

```python
data1
```

```python
import cassiopeia as cas
#print(cas.__version__)
```

```python
cas_tree = cas.data.CassiopeiaTree(character_matrix=data1, 
                                   #cell_meta=description1
                                  )
```

```python
ilp_solver = cas.solver.ILPSolver(
    convergence_time_limit=500,
    maximum_potential_graph_layer_size=500,
    weighted=False,
    seed=1234,
)
ilp_solver.solve(cas_tree)
```
